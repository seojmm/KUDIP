```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_video(path, ms=100):
    
    # Video Capture 객체 생성
    capture = cv2.VideoCapture(path)
    
    # Frame을 저장할 List 선언
    frames = []
    
    while capture.isOpened(): # Video Capture가 준비되었는지 확인
        
        run, frame = capture.read() # 다음 Frame 읽기
        
        if run: # Frame을 읽은 경우
            cv2.imshow("video", frame)
            cv2.waitKey(ms) # Millisecond 단위로 대기
        else: # 재생이 완료되어 더 이상 Frame을 읽을 수 없는 경우
            break
        
        # Frame List에 추가
        frames.append(frame)

    capture.release() # Capture 자원 반납
    cv2.destroyAllWindows() # 창 제거
    
    return np.array(frames, dtype='uint8')
```

```python
import cv2
import numpy as np

def load_video(path, ms=10):
    capture = cv2.VideoCapture(path)
    frames = []
    
    while capture.isOpened():
        run, frame = capture.read()
        if run:
            frames.append(frame)
        else:
            break
        cv2.imshow("video", frame)
        cv2.waitKey(ms)
    capture.release()
    cv2.destroyAllWindows()
    return np.array(frames, dtype='uint8')

def preprocess_frame(frame):
    # 그레이스케일 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 블러링 적용 (노이즈 제거)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # 이진화 처리 (Thresholding)
    _, thresholded = cv2.threshold(blurred, 127, 255, cv2.THRESH_BINARY)
    
    # 엣지 검출 (Canny Edge Detection)
    edges = cv2.Canny(thresholded, 50, 150)
    
    return edges

def detect_head(edges):
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    head_contours = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if 0 < w < 50 and 0 < h < 50:  # 머리 크기 조건
            head_contours.append(cnt)
    return head_contours

def detect_feet(edges):
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    feet_contours = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if 50 < w < 200 and 50 < h < 200:  # 발 크기 조건
            feet_contours.append(cnt)
    return feet_contours

def get_position_and_bounding_box(contours, top=True):
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(largest_contour)
        if top:
            position = (x + w // 2, y)  # 상단 중앙 (머리)
        else:
            position = (x + w // 2, y + h)  # 하단 중앙 (발)
        return position, (x, y, w, h)
    return None, None

def draw_line_and_box(frame, start_point, end_point, head_box, feet_box):
    if start_point and end_point:
        cv2.line(frame, start_point, end_point, (0, 255, 0), 2)  # 초록색 선
    if head_box:
        x, y, w, h = head_box
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)  # 파란색 박스 (머리)
    if feet_box:
        x, y, w, h = feet_box
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)  # 빨간색 박스 (발)

def count_pushups(frames):
    pushup_count = 0
    initial_head_position = None
    is_down = False

    for frame in frames:
        edges = preprocess_frame(frame)
        
        head_contours = detect_head(edges)
        feet_contours = detect_feet(edges)

        head_position, head_box = get_position_and_bounding_box(head_contours, top=True)
        feet_position, feet_box = get_position_and_bounding_box(feet_contours, top=False)

        draw_line_and_box(frame, head_position, feet_position, head_box, feet_box)

        if head_position:
            current_head_y = head_position[1]  # 머리의 y 위치

            if initial_head_position is None:
                initial_head_position = current_head_y

            if current_head_y > initial_head_position + 10 and not is_down:
                is_down = True
            elif current_head_y < initial_head_position - 10 and is_down:
                is_down = False
                pushup_count += 1

            initial_head_position = current_head_y

        cv2.imshow("Processed Video", frame)
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

    cv2.destroyAllWindows()
    return pushup_count

# 비디오 경로 설정
video_path = '/Users/yunjinkim/Desktop/디지털영상처리/video.mov'

# 비디오 로드
video_frames = load_video(video_path)

# 푸쉬업 카운팅
pushup_count = count_pushups(video_frames)
print(f"Total Push-ups Counted: {pushup_count}")
```
