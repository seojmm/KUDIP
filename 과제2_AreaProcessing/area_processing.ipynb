```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from PIL import Image

# img = cv2.imread('./HW3 Image Samples/Noise Filtering/Gaussian noise.png')
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

for i, img in enumerate(glob.glob('./HW3 Image Samples/Noise Filtering/*.jpg')):
	plt.title(os.path.basename(img))
	plt.axis('off')
	img = cv2.imread(img)
	plt.imshow(img, cmap='gray')
	plt.show()
    
	# 1. Gaussian Kernel  
	k1 = np.array([[1, 2, 1],
					[2, 4, 2],
					[1, 2, 1]]) *(1/16)
	blur1 = cv2.filter2D(img, -1, k1)

	# 2. getGaussianKernel
	k2 = cv2.getGaussianKernel(3, 0)
	blur2 = cv2.filter2D(img, -1, k2*k2.T)

	# 3. Gaussian Blur
	blur3 = cv2.GaussianBlur(img, (3, 3), 0)

	#  
	print('k1:', k1)
	print('k2:', k2*k2.T)

	plt.figure(figsize=(10, 4))

	plt.subplot(1, 3, 1)
	plt.axis('off')
	plt.imshow(blur1, cmap='gray')

	plt.subplot(1, 3, 2)
	plt.axis('off')
	plt.imshow(blur2, cmap='gray')

	plt.subplot(1, 3, 3)
	plt.axis('off')
	plt.imshow(blur3, cmap='gray')

	plt.show()
```
